```python
from numpy import *
from matplotlib.pyplot import *

data = loadtxt("A04_data.txt",delimiter=",",skiprows=1)
```

```python
t = data[:,0]*.001
figure()
dist = ((data[:,1]-data[:,3])**2+(data[:,2]-data[:,4])**2)**.5
plot(t,dist,'k')
xlabel('time (s)')
ylabel('distance (m)')


figure()
plot(data[:,1],data[:,2],'k.',data[:,3],data[:,4],'r.')
axis('equal')
```
